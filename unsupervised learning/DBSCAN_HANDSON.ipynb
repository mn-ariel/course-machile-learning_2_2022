```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs

import matplotlib.pyplot as plt
from numpy import random, where

```

```python
x, _ = make_blobs( n_samples=500, centers= 4, cluster_std=0.4)

plt.scatter( x[:, 0], x[:, 1])

plt.show()
```

```python
x.shape
```

```python
x
```

```python
dbscan = DBSCAN( eps = 5.3, min_samples=10 )

print(dbscan)
```

```python
pred = dbscan.fit_predict(x)
```

```python
pred
```

```python
anon_index = where(pred == -1)
cat_one = where(pred==0)
cat_two = where(pred==1)


values = x[anon_index]
values_cat_one = x[cat_one]
values_cat_two = x[cat_two]

```

```python
values
```

```python
plt.scatter(x[:, 0], x[:,1])
plt.scatter(values[:,0], values[:, 1], color = 'r')
plt.scatter(values_cat_one[:,0], values_cat_one[:, 1], color = 'g')
plt.scatter(values_cat_two[:,0], values_cat_two[:, 1], color = 'yellow')

plt.show()
```
